```python
a = [1,3]
b = [1,2,10]
for i in b:
  a.append(i)

def bubble_sort(arr):
    def swap(i, j):
        arr[i], arr[j] = arr[j], arr[i]
    n = len(arr)
    swapped = True  
    x = -1
    while swapped:
        swapped = False
        x = x + 1
        for i in range(1, n-x):
            if arr[i - 1] > arr[i]:
                swap(i - 1, i)
                swapped = True         
    return arr

sorted_arr = bubble_sort(a)
print(sorted_arr)
```

```python
def median(arr):
  leng = len(sorted_arr)
  temp =  int(len(sorted_arr)/2)
  if leng % 2 == 0: 
    return (arr[temp]+arr[temp-1])/2
  else:
    return arr[temp]

median(sorted_arr)
```
